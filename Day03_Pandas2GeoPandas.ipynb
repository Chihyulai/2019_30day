```python
import pandas as pd 
df=pd.read_csv('data/Library/Library.csv',encoding='utf-8')
df
```

```python
import requests,json
def rest(address):
    # YOUR_KEY is needed
    url = 'https://api.tomtom.com/search/2/geocode/{}.json?&key=YOUR_KEY&countrySet=TWN&language=zh-TW&limit=1'.format(
        address)
    response = requests.get(url)
    data = response.text
    js = json.loads(str(data))
    return [js['results'][0]['position']['lon'],js['results'][0]['position']['lat']]
```

```python
import geopandas as gpd
from shapely.geometry import Point
# 這邊使用df.head()只取前五筆測試
geom=[Point(rest(row.address)) for idx, row in df.head(20).iterrows()]
crs = {'init': 'epsg:4326'}
gdf = gpd.GeoDataFrame(df.head(20), crs=crs, geometry=geom)
gdf
```

```python
gdf.plot()
```

```python
gdf.to_file( filename = 'output/Light.shp',encoding='utf-8')
```
