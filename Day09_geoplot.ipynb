```python
import geopandas as gpd
light=gpd.read_file('output/light.shp',encoding='utf-8')
light=light[light.is_valid]
light=light[light['district']=='永和區']
light.crs = {'init' :'epsg:3826'} # 避免資料沒設，這邊再重新給一次
light=light.to_crs(epsg=4326)
light=light.reset_index()
light
```

```python
import geoplot as gplt
gplt.pointplot(light, figsize=(16, 8))
```

```python
import numpy as np
light['year']=np.random.randint(low=0,high=20,size=light.shape[0])
gplt.pointplot(light, hue='year', figsize=(16, 8))
```

```python
import geoplot.crs as gcrs
gplt.pointplot(light, projection=gcrs.AlbersEqualArea(), hue='year', legend=True, scheme='fisher_jenks', figsize=(16, 8))
```

```python
gplt.pointplot(light, projection=gcrs.AlbersEqualArea(), scale='year', legend=True ,limits=(0, 8), figsize=(32, 16))
```

```python
light['size']=np.random.randint(low=0,high=5,size=light.shape[0])
gplt.pointplot(light, projection=gcrs.AlbersEqualArea(),hue='size', scale='year', legend=True ,limits=(0, 8), figsize=(32, 12))
```

```python
village=gpd.read_file('data/village/village.shp',encoding='utf-8')
village=village[village.is_valid]
village=village[village['ADMIT']=='永和區']
village.crs = {'init' :'epsg:3826'} # 避免資料沒設，這邊再重新給一次
village=village.to_crs(epsg=4326)
village=village.reset_index()
gplt.choropleth(village, hue='X', projection=gcrs.AlbersEqualArea(),
        legend=True, edgecolor='white', linewidth=0.5,  scheme='equal_interval',figsize=(16, 8))
```

```python
result=gpd.tools.sjoin(light[['geometry','year','size']], village[['ADMIV','ADMIT','geometry']], op='within',how="right")
result['count']=1
result=result.dissolve(by='ADMIV', aggfunc='sum')
result['year']=result['year']/result['count']
result['size']=result['size']/result['count']
gplt.choropleth(result, hue='count', projection=gcrs.AlbersEqualArea(),
        legend=True, edgecolor='white', linewidth=0.5,  scheme='equal_interval',figsize=(16, 8))
```

```python
import matplotlib.pyplot as plt

ax =gplt.polyplot(village,projection=gcrs.AlbersEqualArea(), figsize=(16, 8))
gplt.pointplot(light,ax=ax,projection=gcrs.AlbersEqualArea(),  hue='year', legend=True)
```
